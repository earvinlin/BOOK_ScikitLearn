## 1. 載入相關套件

```python
from sklearn.datasets import (make_classification, make_blobs, 
                              make_regression, make_circles, make_moons)
```

## 載入分類資料集

```python
X, y = make_classification(n_samples=100, n_classes=3, n_features=20, 
                           n_informative=15, n_redundant=5, random_state=5)
print("X's shape= ", X.shape) # 
print("y's shape= ", y.shape) # 目標變數(y)類別個數
```

```python
print("X= \n", X)
print("y \n", y) 
```

## 繪圖

```python
import matplotlib.pyplot as plt

# 樣本點的形狀
markers = ["x", "o", "^"]

# 針對類別各畫一個散佈圖
for k in range(3):
    X_0 = []
    X_1 = []
    for i in range(len(y)):
        if y[i] == k:
            X_0.append(X[i, 0])
            X_1.append(X[i, 1])
            print("i= ", i, "X_0= ", X_0, "X_1= ", X_1)
            plt.scatter(X_0, X_1, marker=markers[k], s=50)
plt.show()    
```

## 載入集群資料集

```python
X, y, centers = make_blobs(n_samples=100, centers=3, cluster_std=1,
                           n_features=2, return_centers=True)
print(X.shape)
```

```python
centers
```

## 繪圖

```python
import matplotlib.pyplot as plt

# 樣本點的形狀
markers = ["x", "o", "^"]

# 針對類別各畫一個散佈圖
for k in range(3):
    X_0 = []
    X_1 = []
    for i in range(len(y)):
        if y[i] == k:
            X_0.append(X[i, 0])
            X_1.append(X[i, 1])
            plt.scatter(X_0, X_1, marker=markers[k], s=50)

# 繪製集群中心點
X_0 = []
X_1 = []
for i in range(len(centers)):
    X_0.append(centers[i, 0])
    X_1.append(centers[i, 1])
plt.scatter(X_0, X_1, marker='s', s=200, alpha=0.5);
```

## 載入迴歸資料集

```python
X, y, coef = make_regression(n_samples=100, n_features=1, noise=20, coef=True
                             , random_state=123)
print(X.shape)
```

```python
coef
```

## 繪圖

```python
plt.scatter(X[:,0], y)
plt.plot([min(X), max(X)], [min(X)*coef, max(X)*coef], 'r')
```

## 載入非線性的資料集

```python
X, y = make_moons(n_samples=100, noise=0.05)
print(X.shape)
```

## 繪圖

```python
# 針對類別各畫一個散佈圖
colors = ['red', 'blue']
for k in range(2):
    X_0 = []
    X_1 = []
    for i in range(len(y)):
        if y[i] == k:
            X_0.append(X[i, 0])
            X_1.append(X[i, 1])
            plt.scatter(X_0, X_1, s=50, c=colors[k])
plt.show()    
```
